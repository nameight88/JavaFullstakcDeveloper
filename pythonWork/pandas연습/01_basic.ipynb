```python
# 삼성전자 주식 데이터 가져오기
stock_code = '005930'  # 삼성전자
start_date = '2022-01-01'
df = fdr.DataReader(stock_code, start_date)

# 주가 그래프 그리기
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'])
plt.title('삼성전자 주가')
plt.xlabel('날짜')
plt.ylabel('주가')
plt.grid(True)
plt.show()

# 거래량 그래프 그리기
plt.figure(figsize=(12, 6))
plt.bar(df.index, df['Volume'])
plt.title('삼성전자 거래량')
plt.xlabel('날짜')
plt.ylabel('거래량')
plt.grid(True)
plt.show()
```
