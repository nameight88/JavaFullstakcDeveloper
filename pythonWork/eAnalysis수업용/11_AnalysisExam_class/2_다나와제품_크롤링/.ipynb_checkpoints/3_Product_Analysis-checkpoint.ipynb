```python
# 예제 6-34 데이터 불러오기 
import pandas as pd
washing_data = pd.read_excel('./files/crawling_final.xlsx')
washing_data.info()
washing_data.head()

```

```python
# 예제 6-35 세탁용량 기준 정렬
top_list = washing_data.sort_values(["세탁용량"], ascending = False)

# 예제 6-35 건조용량 기준 정렬
top_list = washing_data.sort_values(["건조용량"], ascending = False)

# 흡입력 & 사용시간 TOP 리스트
top_list = washing_data.sort_values(["세탁용량","건조용량"], ascending = False)

# 예제 6-38 평균값 정리
price_mean_value = washing_data['가격'].mean()
suction_mean_value = washing_data['세탁용량'].mean()
use_time_mean_value = washing_data['건조용량'].mean()
#print("가격 평균값", price_mean_value)
#print("세탁 용량 평균값", suction_mean_value)
#print("건조 용량 평균값", use_time_mean_value)

# 예제 6-39 가성비 좋은 제품 탐색
condition_data = washing_data [
    (washing_data['가격'] <= price_mean_value) & 
    (washing_data['세탁용량'] >= suction_mean_value) & 
    (washing_data['건조용량'] >= use_time_mean_value)]
condition_data

# 예제 6-40 라이브러리 임포트 및 한글 글꼴 설정
from matplotlib import font_manager, rc
import matplotlib.pyplot as plt
import seaborn as sns
import platform 
# 그래프에서 한글 표기를 위한 글꼴 변경(윈도우, macOS에 대해 처리)
font_path = ''
if platform.system() == 'Windows': 
    font_path = 'c:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname = font_path).get_name()
    rc('font', family = font_name)
elif platform.system() == 'Darwin':
    font_path = '/Users/$USER/Library/Fonts/AppleGothic.ttf'
    rc('font', family = 'AppleGothic')
else: 
    print('Check your OS system')
    
%matplotlib inline

# 예제 6-41 결측값 없애기
chart_data = washing_data.dropna(axis = 0)
len(chart_data)

# 예제 6-42 흡입력, 사용시간의 최댓값/최솟값 정리
# 흡입력, 사용사간 최대, 최소
suction_max_value = chart_data['세탁용량'].max()
suction_mean_value = chart_data['세탁용량'].mean()
use_time_max_value = chart_data['건조용량'].max()
use_time_mean_value = chart_data['건조용량'].mean()

# 예제 6-43 청소기 성능 시각화
plt.figure(figsize=(20, 10))
plt.title("세탁기 / 건조기 차트")
sns.scatterplot(x = '세탁용량', y = '건조용량', size = 100, hue = chart_data['회사명'], 
            data = chart_data, sizes = (300, 300), legend = False)
plt.plot([0, suction_max_value], 
          [use_time_mean_value, use_time_mean_value], 
          'r--', 
          lw = 1 )
plt.plot([suction_mean_value, suction_mean_value], 
          [0, use_time_max_value], 
          'r--', 
          lw = 1 )
plt.show()




```
