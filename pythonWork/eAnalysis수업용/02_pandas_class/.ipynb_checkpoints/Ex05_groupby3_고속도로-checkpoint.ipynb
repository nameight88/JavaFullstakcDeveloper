## groupby()


#### groupby()와 함께 사용하는 집계 메소드

- count : 누락값을 제외한 데이타 수를 반환
- size : 누락값을 포함한 데이타 수를 반환
- sum
- mean
- std
- var
- min
- max
- quantile(q=0.25)
- quantile(q=0.50)
- quantile(q=0.75)
- describe
- first
- last
- nth   : n번째 행 반환
- sem   : 평균의 표준편차 반환

```python
import pandas as pd

df = pd.read_excel('data/고속버스.xlsx') 
df
```

```python
# 그룹화
df_gruop = df.groupby("선별")
df_gruop
#groupby로 선별이라는 컬럼만 추출해서 변수에 저장
df_gruop.count()
#여기에서 차종만 필요하므로 차종만 추출해서 저장
df_gruop2=df_gruop["차종"].count()
df_gruop2
```

```python
df_gruop2.sort_values(ascending=False)
```

ascending = False 값으로 줘서 내림차순으로 정렬

```python
%matplotlib inline

# 한글처리
from matplotlib import rc
rc('font', family='Malgun Gothic')
#한글처리가 필요한 이유 -> 리눅스에서 그래프로 출력을 할때 한글처리가 안될 수 도 있기 떄문에
```

```python
# 노선별 수에 대한 그래프
df_gruop['선별'].count().plot(kind="bar")
```

```python
df_gruop2.sort_values(ascending=False).plot(kind="bar")
```

```python
# 차종별, 노선별별 그룹화 작업
df[['차종',"선별"]]
temp = df.groupby(['차종','선별']) #차종 중에서도 선별에 대해서 또 분류를 하고 싶을 때 다음과 같이 입력
temp.count()
```

```python
# 차종별, 노선별별 그룹화 작업
df[['차종',"선별"]]
temp = df.groupby(['차종','선별'])
temp2 = temp["총이용인원"].count() #차종중에서도 선별을 분류를 하고 그중에서도 총 이용 인원이 궁금할때
temp2.plot(kind="bar")
#분류를 한 데이터를 그래프로 시각화 작업
```
