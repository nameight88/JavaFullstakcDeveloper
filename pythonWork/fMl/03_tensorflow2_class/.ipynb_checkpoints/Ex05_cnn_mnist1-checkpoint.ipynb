# Fashion MNIST with CNN

```python
# (1) Fashion MNIST 데이터셋 불러오기 및 정규화
import tensorflow as tf

fashion_mnist = tf.keras.datasets.fashion_mnist
(train_X, train_Y), (test_X, test_Y) = fashion_mnist.load_data()

train_X = train_X / 255.0
test_X = test_X / 255.0
```

```python
# (2) 데이터를 채널을 가진 이미지 형태(3차원)으로 바꾸기
#     예전에 Dense 레이어에 훈련 데이타와 테스트 데이터를 통과시켰던 것과 달리
#     Conv2D 레이어로 컨볼루션 연산을 하기 위해 데이타의 shape를 바꾼다.
#     보통 흑백이미지는 1채널, 칼라 이미지는 3채널을 가진다.
#     reshape()를 이용하여 흑백 이미지 채널 1을 가장 뒤쪽에 추가한다.

# reshape 이전
print(train_X.shape, test_X.shape)

train_X = train_X.reshape(-1, 28, 28, 1)
test_X = test_X.reshape(-1, 28, 28, 1)

# reshape 이후
print(train_X.shape, test_X.shape)
```

```python
# 위의 변경된 데이터를 그래프로 확인할 때는 2차원이여야 한다.
import matplotlib.pyplot as plt
# 전체 그래프의 사이즈를 width=10, height=10 으로 지정합니다.
plt.figure(figsize=(10, 10))
for c in range(16):
    # 4행 4열로 지정한 grid 에서 c+1 번째의 칸에 그래프를 그립니다. 1~16 번째 칸을 채우게 됩니다.
    plt.subplot(4,4,c+1)
    plt.imshow(train_X[c].reshape(28,28), cmap='gray')
    
plt.show()

# train 데이터의 첫번째 ~ 16번째 까지의 라벨을 프린트합니다.
print(train_Y[:16])
```

```python
# (3) Fashion MNIST 분류 컨볼루션 신경망 모델 정의
#     모델은 총 3개의 Conv2D 레이어를 사용한다
#     첫 레이어의 input_shape(28,28,1)은 입력이미지높이, 너비, 채널수이다.
#     필터수는 16, 32, 64로 2배씩 증가시킨다
#     Flatten 레이어로 다차원 데이터로 1차원으로 정렬한 다음
#     2개의 Dense 레이어를 사용하여 분류기를 만든다.

model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(input_shape=(28,28,1), kernel_size=(3,3), filters=16),
    tf.keras.layers.Conv2D(kernel_size=(3,3), filters=32),
    tf.keras.layers.Conv2D(kernel_size=(3,3), filters=64),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=128, activation='relu'),
    tf.keras.layers.Dense(units=10, activation='softmax')
])

model.compile(optimizer=tf.keras.optimizers.Adam(), 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])

model.summary()
```

```python
# [참고] Google Colab GPU 사양 확인
# !nvidia-smi

# GPU 아니면 속도 엄청 느림 (30분이상)
```

```python
# (4) Fashion MNIST 분류 컨볼루션 신경망 모델 학습
#    epochs=10으로 줄여서 확인

history = model.fit(train_X, train_Y, epochs=25, validation_split=0.25)
```

```python
# 먼저 여기 출력을 보고 분석한 후에 아래 그래프를 확인한다

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], 'b-', label='loss')
plt.plot(history.history['val_loss'], 'r--', label='val_loss')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], 'g-', label='accuracy')
plt.plot(history.history['val_accuracy'], 'k--', label='val_accuracy')
plt.xlabel('Epoch')
plt.ylim(0.7, 1)
plt.legend()

plt.show()


model.evaluate(test_X, test_Y, verbose=0)
```

[ 결과 ]

    - 왼쪽 그래프를 확인하면 loss는 감소하고 val_loss는 증가하는 정형적인 과적합의 형태이다.
    - 오른쪽 그래프에서는 훈련 데이타에 대한 모델의 정확도 accuracy가 빠르게 증가하지만 검증 데이타에 대한 정확도 val_accuracy는 학습이 진행될수록 감소한다.
    
    - model.evaluate()로 첫번째가 테스트 데이타의 loss이고 두번째가 테스트 데이터의 accurary이다.
        [1.4753566287867725, 0.8525]
    - 테스트 데이터의 accurary는 앞에서 Dense 레이어만 했을 때보다도 낮은 수준이다(확인필요)

## [추가] 모델 개선

[ 아래 ] 레이어 추가로 개선하고자 하는 것인데 기존 코드 복사하며 비교하면서 하자  

    + 과적합을 줄이고자  풀링 레이어, 드랍아웃 레이어 추가

```python
# (3-2) Fashion MNIST 분류 컨볼루션 신경망 모델 정의 - 풀링 레이어, 드랍아웃 레이어 추가
#  
model = tf.keras.Sequential([  
    tf.keras.layers.Conv2D(input_shape=(28,28,1), kernel_size=(3,3), filters=32),  # filters=16 -> filters=32
    tf.keras.layers.MaxPool2D(strides=(2,2)),  # 추가
    tf.keras.layers.Conv2D(kernel_size=(3,3), filters=64),  # filters=32 -> filters=64
    tf.keras.layers.MaxPool2D(strides=(2,2)),  # 추가
    tf.keras.layers.Conv2D(kernel_size=(3,3), filters=128), # filters=64 -> filters=128
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(units=128, activation='relu'),
    tf.keras.layers.Dropout(rate=0.3),  # 추가
    tf.keras.layers.Dense(units=10, activation='softmax')
])

model.compile(optimizer=tf.keras.optimizers.Adam(), 
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])

model.summary()
```

```python
# 위 (4)와 동일
# (4-2) Fashion MNIST 분류 컨볼루션 신경망 모델 학습 - 풀링 레이어, 드랍아웃 레이어 추가
history = model.fit(train_X, train_Y, epochs=25, validation_split=0.25)

import matplotlib.pyplot as plt
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], 'b-', label='loss')
plt.plot(history.history['val_loss'], 'r--', label='val_loss')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], 'g-', label='accuracy')
plt.plot(history.history['val_accuracy'], 'k--', label='val_accuracy')
plt.xlabel('Epoch')
plt.ylim(0.7, 1)
plt.legend()

plt.show()

model.evaluate(test_X, test_Y, verbose=0)
```

[ 결과 ] 

앞에서 테스트 데이터의 정확도가 85% 정도 였는데, 레이어 추가로 89%로 높아졌다.
