## 유방암 데이타셋

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc

# 한글 폰트 설정
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
# 1. 암 데이타셋 로딩


cancer = load_breast_cancer()

print("cancer keys : {}".format(cancer.keys()))
print(cancer.DESCR)
```

```python
# 2.  학습데이타와 검증데이타로 분리 ( 75:25)

X = cancer.data
y = cancer.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
# 3. SVM 모델 생성하고 정확도 출력
svc = svm.SVC()
svc.fit(X_train, y_train)
```

```python
# 4. 검증하기 ( svc.score() 이용)

print('학습용 데이터셋 정확도: {:.3f}'.format(svc.score(X_train, y_train)))
print('검증용 데이터셋 정확도: {:.3f}'.format(svc.score(X_test, y_test)))
```

[ 결과 ]  

검증용 데이터셋이 0.63인데 비해 학습용 데이타셋은 1.0이다.

즉, 과적합화이다

```python
# 과적합화의 원인을 확인하기 위해

plt.boxplot(X_train, manage_xticks=False)

# plt.yscale('symlog') # y축 스케일을 지정할 때와 안할 때의 y축 값 비교
# 한 축만 데이타값이 크기 때문에 기계학습할 때 그 컬럼만 중요한 변수로 인식하여 가중치를 더준다
plt.xlabel('특성목록')
plt.ylabel('특성크기')
```

```python
# cancer.data 데이터셋을 데이터프레임으로 변환
# cancer.feature_names으로 데이터프레임의 컬럼명 지정
df = pd.DataFrame( cancer.data )
df.head()

df.columns = cancer.feature_names
df.head()
```

[ 출력 확인 ]

10단위, 1000단워, 0.1단위로 등등 다른 단위값로 이루어져 기계학습할 때 큰 수에 가중치를 더 부여할 수 있다

그래서, 데이터 전처리로 0~1 사이의 값으로 맞추는 작업이 필요하다


(cf) svc.transform(X_train) 이 있으면 간단할텐데

### 데이타 전처리 필요

```python
#  식 : ( X - min(X) ) / (max(X) - min(X))
# 결과는 항상 0에서 1사이의 값이 된다

min_on_training = X_train.min(axis=0) # axis=0 행방향(컬럼단위(?), axis=1 열방향
# min_on_training[:10]

# 학습용 데이타셋에서 특성별 범위(최대값 - 최소값) 범위 계산
range_on_training = (X_train - min_on_training).max(axis=0)

# 각 특성에 대해 최소값은 0, 최대값은 1로 조정
# 학습용 데이터셋에서 최소값을 빼고 범위로 나누면
X_train_scaled = (X_train - min_on_training)/ range_on_training
X_test_scaled = (X_test - min_on_training) / range_on_training
```

```python
# 확인

print('특성별 최소값\n{}'.format(X_train_scaled.min(axis=0)))
print('특성별 최대값\n{}'.format(X_train_scaled.max(axis=0)))

df2 = pd.DataFrame(X_train_scaled)
df2.head()
```

```python
# SVM 모델에 적용
svc = svm.SVC()
svc.fit(X_train_scaled, y_train)

print('학습용 데이터셋 정확도: {:.3f}'.format(svc.score(X_train_scaled, y_train)))
print('검증용 데이터셋 정확도: {:.3f}'.format(svc.score(X_test_scaled, y_test)))
```

[결과]


- 학습용 데이터셋 정확도: 0.948
- 검증용 데이터셋 정확도: 0.951

데이타 전처리로 데이타 스케일을 조정하니 결과가 크게 달라졌다.

```python
# C값을 증가시키거나 gamma 값을 바꿔서 더 복잡한 모델 만들기
# svc = svm.SVC(C=1)
# svc = svm.SVC(C=100)
svc = svm.SVC(C=1000)

svc.fit(X_train_scaled, y_train)

print('학습용 데이터셋 정확도: {:.3f}'.format(svc.score(X_train_scaled, y_train)))
print('검증용 데이터셋 정확도: {:.3f}'.format(svc.score(X_test_scaled, y_test)))
```

C값 증가로 모델의 성능이 97.2% 까지 향상되었다

## 지도학습에서의 데이터 전처리

```python

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
# 데이타 변환
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 확인
print('변환된 크기 :{}'.format(X_train_scaled.shape))
print('특성별 최소값\n{}'.format(X_train_scaled.min(axis=0)))
print('특성별 최대값\n{}'.format(X_train_scaled.max(axis=0)))

df3 = pd.DataFrame(X_train_scaled)
df3.head()
```

```python
# SVM 모델에 적용
svc = svm.SVC()
svc.fit(X_train_scaled, y_train)

print('학습용 데이터셋 정확도: {:.3f}'.format(svc.score(X_train_scaled, y_train)))
print('검증용 데이터셋 정확도: {:.3f}'.format(svc.score(X_test_scaled, y_test)))
```
