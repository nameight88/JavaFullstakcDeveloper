```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import font_manager, rc

# 한글 폰트 설정
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
##### 유방암 데이타셋 로딩
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()

print("cancer keys : {}".format(cancer.keys()))
```

```python
##### 데이타셋 확인

# 데이타 행과 열의수
print("cancer 행과 열수 : {}".format(cancer.data.shape))

# 클래스별 샘플 갯수 :  악성: 212개, 양성 357개
print('클래스별 샘플 갯수\n{}'.format(
    {n: v for n, v in zip(cancer.target_names, np.bincount(cancer.target))}))

# 특성이름 확인
print('특성 이름: {}'.format(cancer.feature_names))
```

```python
##### 학습용과 검증용 데이타로 분리 

from sklearn.model_selection import train_test_split

X = cancer.data
y = cancer.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, stratify=cancer.target, random_state=0)
            # test_size 생략한 기본값도 0.25 이다 ( 75:25로 구분 )
            # random_state는 난수 발생 seed 값이다
            # stratify는 훈련/테스트 데이터들이 원래의 input dataset의 클래스의 비율과 같은 비율을 가지도록 할 것인지 지정한다. 
            # 예를 들어 0,1의 클래스가 input dataset에 20:80 비율로 있었다면 훈련 데이터와 테스트 데이터 역시 각각의 클래스가 같은 비율로 있도록 지정할 수 있다. 
```

```python
##### 의사결정 트리 (DecisionTree) - 트리모델 생성

from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(random_state=0)

# 학습용 데이타로 기계학습
tree.fit(X_train, y_train)

print('학습용 데이터셋 정확도: {:.3f}'.format(tree.score(X_train, y_train)))
print('검증용 데이터셋 정확도: {:.3f}'.format(tree.score(X_test, y_test)))
```

```python
##### 
tree = DecisionTreeClassifier(max_depth=3, random_state=0)  
    # max_depth의 기본값은 5단계로 생략하면 5단계이다
    # 4 -> 5 - > 3
    # max_depth를 변경하며 최적이 정확도를 찾는다. ( 검증용 정확도가 높은 경우 )

# 학습용 데이타로 기계학습
tree.fit(X_train, y_train)

print('학습용 데이터셋 정확도: {:.3f}'.format(tree.score(X_train, y_train)))
print('검증용 데이터셋 정확도: {:.3f}'.format(tree.score(X_test, y_test)))
```

```python
from sklearn.tree import export_graphviz
import pydotplus
import graphviz
from IPython.display import Image
```

```python

# graphviz의 경로를 환경변수 PATH에 등록
import os
os.environ['PATH'] += os.pathsep + 'C:/Program Files (x86)/Graphviz2.38/bin/'

# class_names['malignant','benign'] : 악성, 양성
dot_data = export_graphviz(tree, out_file=None, feature_names=cancer.feature_names,
                          class_names=cancer.target_names, filled=True, rounded=True, special_characters=True)

# 그래프 생성
graph = pydotplus.graph_from_dot_data(dot_data)
# 그래프를 이미지로 변환
Image(graph.create_png())
```

```python
#####  특성 중요도
#       : 트리 만드는 결정에 각각의 특성(feature)가 얼마나 중요한지를 평가
#       0 ~ 1  사이의 값
#       0 : 전혀 사용되지 않음
#       1 : 타겟 클래스 예측

def plot_feature_cancer(model):
    n_features = cancer.data.shape[1]  # 데이터셋의 필드(컬럼)갯수
    plt.barh(range(n_features), model.feature_importances_, align='center') # 수평막대그래드
    plt.yticks(np.arange(n_features), cancer.feature_names) # 눈금표시
    plt.xlabel('특성중요도')
    plt.ylabel('특성')
    plt.ylim(-1, n_features) # 범위 지정
    
plot_feature_cancer(tree)

# max_depth를 변경하면 중요한 특성이 바뀔수도 있다
```

### 의사결정나무의 단점

    : 과거의 데이터에 대해 정확도는 높지만 새로운 데이타에 대한 예측력이 약함
    
- 과적합화를 방지할 수 있는 대표적인 방법 ==> 랜덤포레스트


* 랜덤포레스트


    - 여러 개의 의사결정나무를 만들고 투표를 통해 다수결로 결과를 결정함
    - 처리가 빠르고 분류 정밀도가 높다
    

```python
from sklearn.ensemble import RandomForestClassifier

# test_size 기본값  75:25
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)


# 100개의 트리(n_estimators)로 이루어진 랜덤 포레스트
# n_estimators 값을 변경하며 최적의 크기값을 찾아야 한다 (10, 100, 1000 확인)
forest = RandomForestClassifier( n_estimators=100, random_state=0)
forest.fit(X_train,y_train)

print('학습용 데이터셋 정확도: {:.3f}'.format(forest.score(X_train, y_train)))
print('검증용 데이터셋 정확도: {:.3f}'.format(forest.score(X_test, y_test)))
```

```python
# 랜덤 포레스트의 각 트리의 특성 중요도
plot_feature_cancer(forest)
```
